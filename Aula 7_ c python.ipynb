```python
numero = int(input("Digite um numero inteiro positivo: "))
fatorial = 1
if numero < 0 :
    print("fatorial não existe!")
elif numero == 0 :
    print("fatorial de 0 é igual a 1.")
else:
    for i in range(1, numero + 1):
        fatorial = fatorial * i
        print(f"fatorial de {numero} é igual a {fatorial}, ")
```
