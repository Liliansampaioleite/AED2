```python
largura = float(input("Digite a largura: "))
cumprimento = float(input("Digite o cumprimento: "))
lata = float(input("Digite o tamanho da lata em litros: "))
area_total = (2 * largura * 2.80) + (2 * cumprimento * 2.80)

area_pintura = area_total - 1.68
litros_necessarios = area_pintura / 3 
quantidade_latas = litros_necessarios / lata 

print(f"Você precisará de {quantidade_latas:.2f} latas ")
```
